Applying Kmeans with 3 clusters, for orders prior for days since prior order.

```python
kmeans_3 = KMeans(n_clusters=3, random_state=0)
```

```python
orderp_km3 = kmeans_3.fit_predict(order_prior.loc[:,['days_since_prior_order','order_number']])
print(orderp_km3)
```

Function defined below to plot results.

```python
def plot_clusters(order_prior, assignment):
    col_dic = {0:'blue',1:'green',2:'orange',3:'gray',4:'magenta',5:'black'}
    colors = [col_dic[x] for x in assignment]
    plt.scatter(order_prior.loc[:,'days_since_prior_order'], order_prior.loc[:,'order_number'], color = colors,)
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Instacart Groceries')
    plt.show()
```

```python
def plot_clusters1(order_prior, assignment):
    col_dic = {0:'blue',1:'green',2:'orange',3:'gray',4:'magenta',5:'black'}
    colors = [col_dic[x] for x in assignment]
    plt.scatter(order_prior.loc[:,'order_hour_of_day'], order_prior.loc[:,'order_number'], color = colors,)
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Instacart Groceries')
    plt.show()
```

```python
plot_clusters(order_prior, orderp_km3)
```
